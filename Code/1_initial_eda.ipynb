```python
# We will import our libraries, set the random seed for reproducibility and load our dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

df = pd.read_csv('../Data/Train.csv')
```

```python
# We will display the basic information about our dataset
print("Dataset Information:")
print(df.info())
print("First few rows of the dataset:")
print(df.head())
```

```python

# We will display our summary statistics
print("Summary Statistics:")
print(df.describe())
```

```python

# We will check for any missing values
missing_values = df.isnull().sum()
print("Missing Values in Each Column:")
print(missing_values[missing_values > 0])
```

```python

# we will perform univariate analysis on the distribution of Lyrics Length
plt.figure(figsize=(10, 6))
sns.histplot(df['len'], bins=30, kde=True, color='blue')
plt.title('Distribution of Lyrics Length')
plt.xlabel('Length of Lyrics')
plt.ylabel('Frequency')
plt.axvline(df['len'].mean(), color='red', linestyle='--', label='Mean Length')
plt.axvline(df['len'].median(), color='green', linestyle='--', label='Median Length')
plt.legend()
plt.savefig('../images/1.1_lyrics_length_distribution.png')
plt.show()
```

```python

# We will plot the distribution of Dating Scores
plt.figure(figsize=(10, 6))
sns.histplot(df['dating'], bins=30, kde=True, color='orange')
plt.title('Distribution of Dating Scores')
plt.xlabel('Dating Score')
plt.ylabel('Frequency')
plt.axvline(df['dating'].mean(), color='red', linestyle='--', label='Mean Dating Score')
plt.axvline(df['dating'].median(), color='green', linestyle='--', label='Median Dating Score')
plt.legend()
plt.savefig('../images/1.2_dating_score_distribution.png')
plt.show()
```

```python

# We will plot the distribution of Romantic Scores
plt.figure(figsize=(10, 6))
sns.histplot(df['romantic'], bins=30, kde=True, color='purple')
plt.title('Distribution of Romantic Scores')
plt.xlabel('Romantic Score')
plt.ylabel('Frequency')
plt.axvline(df['romantic'].mean(), color='red', linestyle='--', label='Mean Romantic Score')
plt.axvline(df['romantic'].median(), color='green', linestyle='--', label='Median Romantic Score')
plt.legend()
plt.savefig('../images/1.3_romantic_score_distribution.png')
plt.show()
```

```python

# We will perform bivariate analysis and do a scatter plot of Dating vs Romantic Scores by Genres that we will categorize
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='dating', y='romantic', hue='genre', style='genre', palette='deep', alpha=0.7)
plt.title('Dating vs Romantic Scores by Genre')
plt.xlabel('Dating Score')
plt.ylabel('Romantic Score')
plt.legend(title='Genre')
plt.savefig('../images/1.4_dating_vs_romantic.png')
plt.show()
```

```python
# We will create a correlation heatmap on selected features from the dataset
plt.figure(figsize=(8, 6))
correlation_matrix = df[['night/time', 'family/gospel', 'movement/places', 'light/visual perceptions']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.savefig('../images/1.5_correlation_heatmap.png')
plt.show()
```

```python

# We will perform multivariate analysis by creating a pairplot of selected features
sns.pairplot(df[['release_date', 'genre', 'lyrics', 'communication']], diag_kind='kde', markers='o')
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.savefig('../images/1.6_pairplot.png')
plt.show()
```

```python

# We will do a boxplot of romantic scores by Genre
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='genre', y='len', palette='Set2')
plt.title('Boxplot of Length by Genre')
plt.xlabel('Genre')
plt.ylabel('Length')
plt.xticks(rotation=45)
plt.savefig('../images/1.7_boxplot_length_by_genre.png')
plt.show()
```
